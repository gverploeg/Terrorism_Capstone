```python
countries = terror[terror['Fatalities'] >= 1].groupby('Country')['Fatalities'].sum().to_frame().reset_index().sort_values('Fatalities', ascending=False)[:15]

countries.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
fig, ax = plt.subplots(1, figsize=(15, 6))
x = countries.iloc[:, 0]
y = countries.iloc[:,1]
ax.bar(x, y, color='b')
ax.set_xlabel("Country", fontsize=14)
ax.set_ylabel("Number of Deaths", fontsize=14)
ax.set_title("Countries with the most Deaths", fontsize=16);
```

```python
mon_deaths = terror[terror['Fatalities'] >= 1].groupby('Month')['Fatalities'].sum().reset_index()

# deaths.head()

import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
fig, ax = plt.subplots(1, figsize=(10, 6))
x = mon_deaths.iloc[:, 0]
y = mon_deaths.iloc[:,1]
ax.bar(x, y, color='b')
ax.set_xlabel("Year", fontsize=14)
xticks = ['', 'January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
# x = np.arange(len(record_counts['genre']))
ax.set_xticks(np.arange(len(xticks)))
ax.set_xticklabels(xticks)
ax.set_ylabel("Number of Deaths", fontsize=14)
ax.set_title("Amount of Deaths By Month", fontsize=16);
```

```python
attack_type = terror.AttackType.value_counts().to_frame().reset_index()
attack_type.columns = ['Attack Type', 'Total Attacks']


import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
fig, ax = plt.subplots(1, figsize=(15, 6))
x = attack_type.iloc[:, 0]
y = attack_type.iloc[:,1]
ax.bar(x, y, color='r')
ax.set_xlabel("Type", fontsize=14)
plt.xticks(rotation=45, fontsize=12, horizontalalignment='right')
ax.set_ylabel("Number of Attacks", fontsize=14)
ax.set_title("Attacks by Type", fontsize=16)
fig.tight_layout();
```

```python
target_type = terror.Target_Type.value_counts().to_frame().reset_index()[:10]
target_type.columns = ['Target Type', 'Total Attacks']


import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
fig, ax = plt.subplots(1, figsize=(15, 6))
x = target_type.iloc[:, 0]
y = target_type.iloc[:,1]
ax.bar(x, y, color='r')
ax.set_xlabel("Target", fontsize=14)
plt.xticks(rotation=45, fontsize=12, horizontalalignment='right')
ax.set_ylabel("Number of Attacks", fontsize=14)
ax.set_title("Attacks by Target", fontsize=16)
fig.tight_layout();
```

```python
players = terror.Group.value_counts().to_frame().reset_index()[:10]
players.columns = ['Group Involved', 'Total Attacks']

import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
fig, ax = plt.subplots(1, figsize=(15, 6))
x = players.iloc[:, 0]
y = players.iloc[:,1]
ax.bar(x, y, color='r')
ax.set_xlabel("Group", fontsize=14)
plt.xticks(rotation=45, fontsize=12, horizontalalignment='right')
ax.set_ylabel("Number of Attacks", fontsize=14)
ax.set_title("Attacks by Group", fontsize=16)
fig.tight_layout();
```
